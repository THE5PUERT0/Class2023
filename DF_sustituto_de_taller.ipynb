```python
#eliminar columna N/A. Luego me muestra el resultado
cunr = cunr.drop(['Unnamed: 26'],axis=1) 
cunr.head()
```

```python
# agrupo columnas y le pido que me saque el promedio 
cunr.groupby(['Grado', 'Sexo', 'Edad'])[['TotalMicro', 'TotalMacro', 'Total']].mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(cunr['Total'], bins=30, alpha=0.3, color='green')
# nombrar los títulos de los ejes y el título del plot
plt.xlabel('Total')
# que me muestre la gráfica
plt.show()
```

```python
#con r le digo que me agrupe por grado y saque el promedio
%%R
prom_edad_grado <- cunr%>%
      group_by(Grado)%>%
      summarize(Edad = mean(Edad))
```

```python
#que me muestre el nuevo df llamado
%%R
prom_edad_grado
```

```python
#que agrupe las 3 columnas, y me muestre los 3 promedios de totales
%%R
tabla <- cunr %>%
group_by(Grado, Sexo, Edad) %>%
summarize(
    PromedioTotalMicro = mean(TotalMicro),
    PromedioTotalMacro = mean(TotalMacro),
    Promedio = mean(Total)    
)
```
